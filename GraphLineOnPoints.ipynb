```python
import random
from matplotlib import pyplot as plt
import numpy as np
import pandas
```

```python
points = pandas.read_csv('output.csv')
X = [0, 4.5]
Y = [0, 9]
plt.scatter(points['Meetings Attended'], points['Grade Increase'])
plt.plot(X, Y)
plt.show()
```

Calculate the sum of squared residuals:

```python
m = -2
sum = 0
iteration = 0
for x in points['Meetings Attended']:
    y = m * x
    sum += ((y - points.loc[iteration, 'Grade Increase']) ** 2)
    iteration += 1
print(sum)
```

Plot guesses for m against cost

```python
plt.cla()
m = -2
while m < 2:
    sum = 0
    iteration = 0
    for x in points['Meetings Attended']:
        y = m * x
        sum += ((y - points.loc[iteration, 'Grade Increase']) ** 2)
        iteration += 1
    plt.scatter(m, sum, color='red')
    m += .1
plt.show()
```
